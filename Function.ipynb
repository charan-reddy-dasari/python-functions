1.  What is the difference between a function and a method in Python?
  - Function: A standalone block of code defined using the def keyword that performs a specific task. Functions can be called independently.
  - def func_name():
    - print("charan reddy")
  - Method: A function that is associated with an object and can access the object's data. Methods are called on objects and are defined inside classes.
  - class class_name:
    - def method_name(self):
      - print("charan reddy")
  - c = class_name()
  - c.method_name()

2.  What is the difference between a function and a method in Python?

  - Argument: The actual value passed to a function when it is called.
  - def func_name(a,b): // a,b are paramters
    - return a+b
  - func_name(10,5) // 10,5 are arguments

5. What are iterators in Python and how do they differ from iterables?
  - Iterable: An object that can be iterated over (e.g., lists, strings, tuples). It has an __iter__() method.

  - l = [a,b,c] // l is iterable
  - it = iter(l) // it is iterator
  - next(it)

```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def func1(l):
  ans = 0
  for i in l:
    if i%2 == 0:
      ans = ans+i;
  return ans
func1([1,2,3,4,5,6,7,8])

```

```python
# 2. Create a Python function that accepts a string and returns the reverse of that string.
def func2(s):
  l = list(s)
  for i,j in zip(range(0,len(s)), range(len(s)-1,-1,-1)):
    if i<j:
      l[i],l[j] = l[j],l[i]
  return "".join(l)
func2("pwskills")
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
import math
def func(num):
  for i in range(2,int(math.sqrt(num))):
    if num % i == 0:
      print("number is not prime")
  else:
    print("number is prime")

num = int(input("enter number : "))
func(num)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def func(s):
  for i in s:
    if i not in ['a','e','i','o','u']:
      return i
"".join(list(filter(func, "pwskills")))
```

```python
# 8.  Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
l = [(1, 5), (3, 2), (7, 8), (4, 3)]

s_list = sorted(l, key=lambda x: x[1])

s_list
```

```python
# 6.  Write a generator function in Python that yields the powers of 2 up to a given exponent.

def func4(n):
  for i in range(n+1):
    yield 2**i
list(func4(10))
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms

fib = lambda x : x if x < 2 else fib(x-1)+fib(x-2)
series = map(fib, [i for i in range(10)])
next(series)
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()
```
